<h1>Testing Correlation using One way Anova , Analysis of Variance </h1>

```python
from scipy.stats import f_oneway
```

<h1>Syntax Requirement</h1>
    <ul>
        <li>specify the value for categorical</li>
        <li>specify the quantitative data </li>
    </ul>

```python
#male
#female

f_stats, p_value = f_oneway(early_bird[early_bird['sex'] == 'male']['body_mass_g'],
                            early_bird[early_bird['sex'] == 'female']['body_mass_g'])

print(p_value)
```
